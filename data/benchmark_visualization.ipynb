# Cost Structure Grouped Bar Chart (Altair)


```python
# Load the data
df = pd.read_csv("saas_financial_market_dataset_with_country.csv")

# Ensure that 'Year' is treated as a categorical variable
df["Year"] = df["Year"].astype(str)

# For each year, find the top 5 companies by Expenses_USD in that year
top5_each_year = (
    df.sort_values(['Year', 'Expenses_USD'], ascending=[True, False])
      .groupby('Year')
      .head(5)
)

# Only keep the exact (Company, Year) pairs present in top5_each_year (no extra companies in any year's bars)
bar_data = top5_each_year[["Company", "Year", "Expenses_USD"]].copy()

# Reset index to avoid any Altair issues
bar_data = bar_data.reset_index(drop=True)

# Create one chart per year, so names only appear for the given year
year_charts = []
for year, group in bar_data.groupby("Year"):
    chart = alt.Chart(group).mark_bar().encode(
        x=alt.X("Company:N", title="Company"),
        y=alt.Y("Expenses_USD:Q", title="Expenses (USD)"),
        color=alt.Color("Company:N", title="Company"),
        tooltip=[
            alt.Tooltip("Company", title="Company"),
            alt.Tooltip("Year", title="Year"),
            alt.Tooltip("Expenses_USD", title="Expenses (USD)", format=",.2f")
        ]
    ).properties(
        title=f"Top 5 Companies by Expenses in {year}",
        width=300,
        height=400
    )
    year_charts.append(chart)

# Display all charts, one below another
cost_structure_charts = alt.vconcat(*year_charts).resolve_scale(color='independent')
cost_structure_charts
```
